```python
import pandas as pd
```

```python
# Total
europe_worker_force_total = pd.read_excel('europe_worker_force.xlsx', 'Sheet 1', dtype=str)

# Managers
europe_worker_force_managers = pd.read_excel('europe_worker_force.xlsx', 'Sheet 2', dtype=str)

# Professionals
europe_worker_force_professionals = pd.read_excel('europe_worker_force.xlsx', 'Sheet 3', dtype=str)

# Technicians and associate professionals
europe_worker_force_technicians = pd.read_excel('europe_worker_force.xlsx', 'Sheet 4', dtype=str)

# Clerical support workers
europe_worker_force_technicians_clerical_support = pd.read_excel('europe_worker_force.xlsx', 'Sheet 5', dtype=str)
```

```python
def clean_database(df):
    
    df = df.drop(range(0, 11))
    df = df.drop(range(50, 57))

    columns = df.columns
    columns = columns.tolist()
    for index, column in enumerate(columns.copy()):
        if index == 0:
            continue
        elif index % 2 == 0:
            columns.remove(column)
    df = df[columns]

    columns_dict_keys = (df.columns).tolist()
    columns_dict_values = ['Country','2019_total', '2019_males', '2019_females', '2020_Q1_total', '2020_Q1_males', '2020_Q1_females', '2020_Q2_total', '2020_Q2_males', '2020_Q2_females', '2020_Q3_total', '2020_Q3_males', '2020_Q3_females', '2020_Q4_total', '2020_Q4_males', '2020_Q4_females', '2021_Q1_total', '2021_Q1_males', '2021_Q1_females', '2021_Q2_total', '2021_Q2_males', '2021_Q2_females', '2021_Q3_total', '2021_Q3_males', '2021_Q3_females', '2021_Q4_total', '2021_Q4_males', '2021_Q4_females', '2022_total', '2022_males', '2022_females']
    columns_rename_dict = dict(zip(columns_dict_keys, columns_dict_values))
    df.rename(columns=columns_rename_dict, inplace=True)
    
    df[columns_dict_values] = df[columns_dict_values].replace(':', '', regex = True)

    for index, column in enumerate(columns_dict_values):
        if index > 0:
            df[column] = pd.to_numeric(df[column],errors = 'coerce')

    # df['mean_2020_male'] = df.iloc[5,8,11,14].mean(axis=1)
    df['2020_males'] = df[['2020_Q1_males', '2020_Q2_males', '2020_Q3_males', '2020_Q4_males']].mean(axis=1)
    df['2020_females'] = df[['2020_Q1_females', '2020_Q2_females', '2020_Q3_females', '2020_Q4_females']].mean(axis=1)
    df['2021_males'] = df[['2021_Q1_males', '2021_Q1_males', '2021_Q1_males', '2021_Q1_males']].mean(axis=1)
    df['2021_females'] = df[['2021_Q1_females', '2021_Q1_females', '2021_Q1_females', '2021_Q1_females']].mean(axis=1)
    df['2020_total'] = df['2020_males'] + df['2020_females']
    df['2021_total'] = df['2021_males'] + df['2021_females']

    df = df.drop(df.iloc[:, 4:28],axis = 1)

    df = df[['Country','2019_total', '2019_males', '2019_females', '2020_total', '2020_males', '2020_females', '2021_total', '2021_males', '2021_females', '2022_total', '2022_males', '2022_females']]
   
    df.reset_index(drop=True, inplace=True)

    return df
```

```python
europe_worker_force_total = clean_database(europe_worker_force_total)
europe_worker_force_managers = clean_database(europe_worker_force_managers)
europe_worker_force_professionals = clean_database(europe_worker_force_professionals)
europe_worker_force_technicians = clean_database(europe_worker_force_technicians)
europe_worker_force_technicians_clerical_support = clean_database(europe_worker_force_technicians_clerical_support)
```

```python
europe_worker_force_managers.drop(range(39, 40))
europe_worker_force_technicians_clerical_support.drop(range(39, 40))
```

```python
europe_worker_force_total.to_csv('total_european_work_force.csv')
europe_worker_force_managers.to_csv('managers_european_work_force.csv')
europe_worker_force_professionals.to_csv('professionals_european_work_force.csv')
europe_worker_force_technicians.to_csv('technicians_european_work_force.csv')
europe_worker_force_technicians_clerical_support.to_csv('support_european_work_force.csv')
```
