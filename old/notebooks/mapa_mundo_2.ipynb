```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
import numpy as np
from matplotlib.gridspec import GridSpec
import matplotlib as mpl
from funciones_data_visualization import *
from variables_data_visualization import *
from pylab import rcParams
import geopandas as gpd
import contextily as ctx
import folium
```

```python
indicators_world_female_shares = pd.read_csv('indicators_world_female_shares.csv')
```

```python
indicators_world_female_shares = indicators_world_female_shares[['Country', 'Life satisfaction']]
```

```python
world_2 = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
new_table_world = world_2.merge(indicators_world_female_shares, how="left", left_on=['name'], right_on=['Country'])
new_table_world = new_table_world.dropna(subset=['Life satisfaction'])
```

```python
my_map_world = folium.Map()
```

```python
folium.Choropleth(
    geo_data=new_table_world,
    name='choropleth',
    data=new_table_world,
    columns=['Country', 'Life satisfaction'],
    key_on='feature.properties.name',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Percentage of women in boards'
).add_to(my_map_world)
my_map_world.save('mapa_mundo_life_satisfaction.html')
```

```python
style_function = lambda x: {'fillColor': '#ffffff', 
                            'color':'#000000', 
                            'fillOpacity': 0.1, 
                            'weight': 0.1}
highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.50, 
                                'weight': 0.1}
NIL = folium.features.GeoJson(
    new_table_world,
    style_function=style_function, 
    control=False,
    highlight_function=highlight_function, 
    tooltip=folium.features.GeoJsonTooltip(
        fields=['Country', 'Life satisfaction'],
        aliases=['Pais: ','Life satisfaction: '],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;") 
    )
)
my_map_world.add_child(NIL)
my_map_world.keep_in_front(NIL)
folium.LayerControl().add_to(my_map_world)
my_map_world.save('mapa_mundo_life_satisfaction.html')
```
