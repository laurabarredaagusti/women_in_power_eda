```python
import pandas as pd
```

```python
satisfaction = pd.read_csv('satisfaction.csv')
```

```python
satisfaction = satisfaction[['Country', 'Indicator', 'Value']]
```

```python
quality_of_support_network = satisfaction['Indicator'] == 'Quality of support network'
educational_attainment = satisfaction['Indicator'] == 'Educational attainment'
years_education = satisfaction['Indicator'] == 'Years in education'
life_satisfaction = satisfaction['Indicator'] == 'Life satisfaction'
```

```python
quality_of_support_network = satisfaction[quality_of_support_network].rename(columns={'Value' : 'Quality of support network'})
educational_attainment = satisfaction[educational_attainment].rename(columns={'Value' : 'Educational Attainment'})
years_education = satisfaction[years_education].rename(columns={'Value' : 'Years in education'})
life_satisfaction = satisfaction[life_satisfaction].rename(columns={'Value' : 'Life satisfaction'})
```

```python
quality_of_support_network = quality_of_support_network[['Country', 'Quality of support network']]
educational_attainment = educational_attainment[['Country', 'Educational Attainment']]
years_education = years_education[['Country', 'Years in education']]
life_satisfaction = life_satisfaction[['Country', 'Life satisfaction']]
```

```python
merged_indicators = pd.merge(quality_of_support_network, educational_attainment, on='Country')
```

```python
merged_indicators = pd.merge(merged_indicators, years_education, on='Country')
```

```python
merged_indicators = pd.merge(merged_indicators, life_satisfaction, on='Country')
```

```python
merged_indicators.drop_duplicates(subset=['Country'], keep='first', inplace=True, ignore_index=True)
merged_indicators
```

```python
share_boards = pd.read_csv('Female share of seats on boards of the largest publicly listed companies.csv')
```

```python
shared_boards_2021 = share_boards['Time'] == 2021
```

```python
shared_boards_2021 = share_boards[shared_boards_2021]
shared_boards_2021 = shared_boards_2021[['Country', 'Value']]
```

```python
merged_everything = pd.merge(merged_indicators, shared_boards_2021, on='Country')
```

```python
result = merged_everything.dtypes
print(result)
```
