```python
import pandas as pd
import numpy as np
```

```python
women_boards_europe_detail = pd.read_excel('women_boards_europe.xlsx', 'women_boards_detail')
women_boards_europe_total = pd.read_excel('women_boards_europe.xlsx', 'women_boards_total')
```

```python
women_boards_europe_detail
```

```python
columns_details = {'time': 'year', 'geo' : 'country', 'value' : 'number', 'POSITION' : 'position',
                    '_POSITION' : 'position init', '_sex' : 'sex init', 'NACE' : 'nace', '_NACE' : 'nace init'
                } 
```

```python
women_boards_europe_detail
```

```python
women_boards_europe_detail = women_boards_europe_detail.rename(columns = columns_details)
```

```python
women_boards_europe_detail = women_boards_europe_detail[['year', 'country', 'number', 'position', 'position init', 'sex', 'nace', 'nace init']]
```

```python
women_boards_europe_detail
```

```python
women_boards_europe_total = women_boards_europe_total.drop(range(0,17))
women_boards_europe_total.columns = women_boards_europe_total.iloc[0]
women_boards_europe_total = women_boards_europe_total.iloc[1: , :]
women_boards_europe_total.set_index('Row Labels',inplace=True)
women_boards_europe_total.index.name = None
```

```python
women_boards_europe_total = women_boards_europe_total.replace(['NaN'],np.nan)
women_boards_europe_total = women_boards_europe_total.apply (pd.to_numeric, errors='coerce')
women_boards_europe_total = women_boards_europe_total.dropna()
women_boards_europe_total
```

```python
women_boards_europe_detail.to_csv('women_boards_europe_detail.csv')
women_boards_europe_total.to_csv('women_boards_europe_total.csv')
```

```python
women_boards_europe_detail
# women_boards_europe_detail.column.str.replace("(::).*","")
```

```python
women_boards_europe_detail_women = women_boards_europe_detail.loc[women_boards_europe_detail['sex'] == 'Women']

women_boards_europe_detail_total = women_boards_europe_detail.loc[women_boards_europe_detail['sex'] == 'Total']

list_of_total = women_boards_europe_detail_total['number'].to_list()
```

```python
women_boards_europe_detail_women.loc[:, "Total"] = list_of_total
women_boards_europe_detail_women
women_boards_europe_detail_women['year'] = women_boards_europe_detail_women['year'].apply(lambda x: x.split('-')[0])
result = women_boards_europe_detail_women.dtypes
print(result)
```

```python
women_boards_europe_detail_women['year'].astype(str).astype(int)
```

```python
percentage = 100 * women_boards_europe_detail_women["number"] / women_boards_europe_detail_women["Total"]
women_boards_europe_detail_women["percentage"] = percentage
women_boards_europe_detail_women.to_csv('women_boards_europe_women_percentage.csv')
```
