```python
import pandas as pd
import numpy as np
```

```python
women_boards_worldwide = pd.read_csv('Female share of seats on boards of the largest publicly listed companies.csv')
```

```python
women_boards_worldwide = women_boards_worldwide[['Country', 'Time', 'Value']]
women_boards_worldwide.sort_values(by=['Country'], inplace=True)
women_boards_worldwide.reset_index(drop=True, inplace=True)
women_boards_worldwide
```

```python
country_list = women_boards_worldwide.Country.unique()
country_list = country_list.tolist()
print(country_list)
year_column_list = women_boards_worldwide.Time.unique()
year_column_list = sorted(year_column_list)
print(year_column_list)
```

```python
def get_year_values(year, country_list):
    year_list = []
    for country in country_list:
        try:
            value = women_boards_worldwide[(women_boards_worldwide['Country'] == str(country)) & (women_boards_worldwide['Time'] == year)][['Value']]
            value_ = (value.values[0])[0]
            year_list.append(value_)
        except:
            year_list.append('NaaN')
    return year_list
```

```python
def get_columns_dictionary(year_column_list,country_list):
    column_dictionary = {}
    column_dictionary['Country'] = country_list
    for year in year_column_list:
        if str(year) not in column_dictionary.keys():
            column_dictionary[str(year)] = get_year_values(year, country_list)
        else:
            column_dictionary[str(year)].append(get_year_values(year, country_list))

    return column_dictionary
```

```python
years_dictionary = get_columns_dictionary(year_column_list,country_list)
```

```python
women_boards_worldwide_dataframe = pd.DataFrame.from_dict(years_dictionary)
```
