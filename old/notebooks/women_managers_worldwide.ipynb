```python
import pandas as pd
import numpy as np
```

```python
women_managers_worldwide = pd.read_csv('Share of female managers.csv')
```

```python
pd.set_option('max_columns', None)
women_managers_worldwide = women_managers_worldwide[['Country', 'Time', 'Value']]
pd.set_option('max_columns', None)
women_managers_worldwide
```

```python
women_managers_worldwide.to_csv('all_women_managers_percentage_worldwide.csv')
```

```python
women_managers_worldwide.sort_values(by=['Country'], inplace=True)
country_list = women_managers_worldwide.Country.unique()
country_list = country_list.tolist()
print(country_list)
year_column_list = women_managers_worldwide.Time.unique()
year_column_list = sorted(year_column_list)
print(year_column_list)
```

```python
def get_year_values(year, country_list):
    year_list = []
    for country in country_list:
        try:
            value = women_managers_worldwide[(women_managers_worldwide['Country'] == str(country)) & (women_managers_worldwide['Time'] == year)][['Value']]
            value_ = (value.values[0])[0]
            year_list.append(value_)
        except:
            year_list.append('NaaN')
    return year_list
```

```python
def get_columns_dictionary(year_column_list,country_list):
    column_dictionary = {}
    column_dictionary['Country'] = country_list
    for year in year_column_list:
        if str(year) not in column_dictionary.keys():
            column_dictionary[str(year)] = get_year_values(year, country_list)
        else:
            column_dictionary[str(year)].append(get_year_values(year, country_list))

    return column_dictionary
```

```python
years_dictionary_managers = get_columns_dictionary(year_column_list,country_list)
```

```python
women_managers_worldwide_dataframe = pd.DataFrame.from_dict(years_dictionary_managers)
women_managers_worldwide_dataframe
```

```python
women_managers_worldwide_dataframe = women_managers_worldwide_dataframe.replace(['NaaN'],np.NaN)
women_managers_worldwide_dataframe.loc[women_managers_worldwide_dataframe['2021'].isnull(),'2021'] = women_managers_worldwide_dataframe['2020']
women_managers_worldwide_dataframe.loc[women_managers_worldwide_dataframe['2021'].isnull(),'2021'] = women_managers_worldwide_dataframe['2019']
women_managers_worldwide_dataframe.loc[women_managers_worldwide_dataframe['2021'].isnull(),'2021'] = women_managers_worldwide_dataframe['2018']
women_managers_worldwide_dataframe.loc[women_managers_worldwide_dataframe['2021'].isnull(),'2021'] = women_managers_worldwide_dataframe['2017']

women_managers_worldwide_dataframe.sort_values(by=['2021'])
```

```python
paises_seleccionados = [15, 11, 37, 40, 12, 0, 35, 4, 36, 18, 41, 30, 16, 3, 21, 32, 9, 22]
paises_seleccionados_2021 = women_managers_worldwide_dataframe.iloc[women_managers_worldwide_dataframe.index.isin(paises_seleccionados)]

paises_seleccionados_2021 = women_managers_worldwide_dataframe.iloc[women_managers_worldwide_dataframe.index.isin(paises_seleccionados)]
paises_seleccionados_2021 = paises_seleccionados_2021[['Country', '2021']]
paises_seleccionados_2021.rename(columns = {'2021':'females'}, inplace = True)
paises_seleccionados_2021["males"] = np.nan
paises_seleccionados_2021.reset_index(drop=True, inplace=True)
paises_seleccionados_2021

for i, row in paises_seleccionados_2021.iterrows(): 
    paises_seleccionados_2021['males'][i] = 100 - paises_seleccionados_2021['females'][i]

paises_seleccionados_2021
paises_seleccionados_2021.to_csv('paises_seleccionados_managers_world_percentage.csv')
# paises_seleccionados_2021.rename(columns = {'2021':'females'}, inplace = True)
# paises_seleccionados_2021["males"] = np.nan
# paises_seleccionados_2021.reset_index(drop=True, inplace=True)
# paises_seleccionados_2021

# for i, row in paises_seleccionados_2021.iterrows(): 
#     paises_seleccionados_2021['males'][i] = 100 - paises_seleccionados_2021['females'][i]

# paises_seleccionados_2021
# paises_seleccionados_2021.to_csv('paises_seleccionados_board_world_percentage.csv')
```
